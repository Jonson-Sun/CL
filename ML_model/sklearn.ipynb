```python
from sklearn import datasets
iris=datasets.load_iris()

from sklearn import decomposition as dcpt
pca=dcpt.PCA(n_components=3)
iris_pca=pca.fit_transform(iris.data)
print(iris_pca[:5])
print(pca.explained_variance_ratio_.sum())
print(iris.data.shape,"^^^",iris_pca.shape)
```

```python
from sklearn import datasets
boston=datasets.load_boston()
from sklearn.gaussian_process import GaussianProcessRegressor

X=boston.data
Y=boston.target
trainset=np.random.choice([True,False],len(Y),p=[.75, .25])
gaus_p=GaussianProcessRegressor()
gaus_p.fit(X[trainset],Y[trainset])
```

```python
from sklearn import datasets
boston=datasets.load_boston()
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(boston.data,boston.target)
result=lr.predict(boston.data)
print(result[:20])
print(lr.coef_)

def MSE_1(target,predict):
    return np.mean(np.power(target-predict,2)  )
    
print(MSE_1(boston.target,result) )

from sklearn.cluster import KMeans
km6=KMeans(n_clusters=6)
km6.fit(boston.data)
#km6.cluster_centers_

from matplotlib  import pyplot as plt
f,ax=plt.subplots(figsize=(7.5,7.5))
ax.scatter(boston.data[:,2],boston.data[:,1])
ax.legend(loc='best')

f,ax=plt.subplots(figsize=(10,5))
ax.hist(boston.target)
```
